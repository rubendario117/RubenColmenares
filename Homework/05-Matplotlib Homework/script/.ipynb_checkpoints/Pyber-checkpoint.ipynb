```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read Files
city_data_to_load = "../data/city_data.csv"
ride_data_to_load = "../data/ride_data.csv"

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

city_df.head()
```

```python
pyber_df = pd.merge(city_df, ride_df, on="city")
pyber_df.head()
```

```python
pyber_grouped = pyber_df.groupby('city')

avg_per_city = pyber_grouped['fare'].mean()
#avg_per_city

total_rides = pyber_grouped['ride_id'].count()
#total_rides

drivers = pyber_grouped['driver_count'].sum()
#drivers

#city_type = pyber_grouped['type']
city_type = city_df.set_index('city')['type']

pyber_dataframe = pd.DataFrame({
    "Total Number of Rides": total_rides,
    "Average Fare per City": avg_per_city,
    "Total Number of Drivers": drivers,
    "City Type": city_type,
})

pyber_dataframe.head()
```

```python
#Create DF per type of city:
urban = pyber_dataframe[pyber_dataframe['City Type'] == 'Urban']
suburban = pyber_dataframe[pyber_dataframe['City Type'] == 'Suburban']
rural = pyber_dataframe[pyber_dataframe['City Type'] == 'Rural']

x_limit = pyber_dataframe['Total Number of Rides'].max()
y_limit = pyber_dataframe['Average Fare per City'].max()

#print scatter plot
plt.title('Pyber Ride Sharing Data')
plt.xlabel('Number of Rides per City')
plt.ylabel('Average Fare per City')


plt.ylim(10, y_limit+5)


plt.xlim(0, x_limit+5)

plt.scatter(urban['Total Number of Rides'], urban['Average Fare per City'], marker="o", s=urban['Total Number of Drivers'], facecolors="lightcoral", edgecolors="black",
           label='Urban',alpha=0.75)

plt.scatter(suburban['Total Number of Rides'], suburban['Average Fare per City'], marker="o",s=suburban['Total Number of Drivers'], facecolors="lightskyblue", edgecolors="black",
           label='Suburban',alpha=0.75)

plt.scatter(rural['Total Number of Rides'], rural['Average Fare per City'], marker="o",s=rural['Total Number of Drivers'], facecolors="gold", edgecolors="black",
           label='Rural',alpha=0.75)


lgnd = plt.legend(frameon = True, edgecolor = 'black')
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]


plt.legend(loc='upper right')
plt.show()
```

Type Percent

```python
#% of Total Fares by City Type
city_by_type = pyber_df.groupby('type')

total_fare = city_by_type.sum()['fare']


labels = ['Urban','Suburban','Rural']
colors = ['lightcoral',"lightskyblue","gold"]

explode = [0, 0, .3]

plt.pie(total_fare, labels=labels, colors=colors, explode = explode,
        autopct="%1.1f%%", shadow=True, startangle=120)

plt.title('% of Total Fares by City Type')
plt.axis('equal')
plt.show()
```

```python
#% of Total Rides by City Type
percentage_urban = (urban['Total Number of Rides'].sum()/pyber_dataframe['Total Number of Rides'].sum())*100
percentage_suburban = (suburban['Total Number of Rides'].sum()/pyber_dataframe['Total Number of Rides'].sum())*100
percentage_rural = (rural['Total Number of Rides'].sum()/pyber_dataframe['Total Number of Rides'].sum())*100

labels = ['Urban','Suburban','Rural']
colors = ['lightcoral',"lightskyblue","gold"]

explode = [0, 0, .3]

total_rides_percentage = [percentage_urban,percentage_suburban,percentage_rural]

plt.pie(total_rides_percentage, labels=labels, colors=colors, explode = explode,
        autopct="%1.1f%%", shadow=True, startangle=120)

plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.show()
```

```python
#% of Total Drivers by City Type
percentage_urban = (urban['Total Number of Drivers'].sum()/city_df['driver_count'].sum())*100
percentage_suburban = (suburban['Total Number of Drivers'].sum()/city_df['driver_count'].sum())*100
percentage_rural = (rural['Total Number of Drivers'].sum()/city_df['driver_count'].sum())*100

labels = ['Urban','Suburban','Rural']
colors = ['lightcoral',"lightskyblue","gold"]

total_rides_percentage = [percentage_urban,percentage_suburban,percentage_rural]

plt.pie(total_rides_percentage, labels=labels, colors=colors,explode = explode,
        autopct="%1.1f%%", shadow=True, startangle=150)

plt.title('% of Total Drivers by City Type')
plt.show()
```

# Analysis:

We can see through the pie charts that the areas with most drivers and rides are on the cities (urban and suburban locations). However, the area with the higher fares are on the rural locations. This could be because in rural areas the distances are way longer than in city ones, therefore rural's rides take longer and become more expensive. We can confirm this assumption by seeing that the lowest fares are from urban areas.
