```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from citipy import citipy
from api_keys import api_key
```

```python
#Create dataframe for cities and its coordinates
cities_df = pd.DataFrame(columns=['City', "Country"])

#Fill the dataframe with random coordinates and cities:
for x in range(1500):
    lat = np.random.uniform (low=-90.000,high=90.000)
    lng = np.random.uniform (low=-180.000,high=180.000)
    city = citipy.nearest_city(lat, lng)
    cities_df = cities_df.append({"City": city.city_name,"Country": city.country_code,}, ignore_index=True)

cities_df.head()
cities_df.count()
```

# Let's start with the API Calls:

```python
#Build the target URL for API Calls:

url = "http://api.openweathermap.org/data/2.5/weather?"
#target_url = url + "appid=" + api_key + "&units=IMPERIAL" + "&q="

#Create a new dataframe for our API calls' results"
weather_df = pd.DataFrame(columns=["City", "Country","Temperature",
                "Latitude","Longitude","Humidity (%)",
                "Cloudiness (%)", "Wind Speed"])

#Start the API calls for all cities in our dataframe:
for index, row in cities_df.iterrows():
    target_city = row['City']
    target_url = url + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + target_city
    city_response = requests.get(target_url).json()

    #Lets fill our cities_df
    try:
        print(f"Querying {target_city}.")
        
        weather_df.loc[index, 'City'] = city_response["name"]
        weather_df.loc[index, 'Country'] = city_response["sys"]['country']
        weather_df.loc[index, 'Temperature'] = city_response["main"]["temp"]
        weather_df.loc[index, 'Latitude'] = city_response["coord"]["lat"]
        weather_df.loc[index, 'Longitude'] = city_response["coord"]["lon"]
        weather_df.loc[index, 'Humidity (%)'] = city_response["main"]["humidity"]
        weather_df.loc[index, 'Cloudiness (%)'] = city_response["clouds"]["all"]      
        weather_df.loc[index, 'Wind Speed'] = city_response["wind"]["speed"]      

    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
    
```

```python
weather_df.head()
```

```python
#Rename column indexes:
weather_df.rename(columns={'Temperature':'Temperature (F)', 'Wind Speed':'Wind Speed (mph)'}, inplace=True)

# Remove duplicate cities from the DF 
weather_df.sort_values("Latitude", inplace=True) 
weather_df.drop_duplicates(keep=False,inplace=True)
weather_df.reset_index()

#Write DF to CSV
weather_df.to_csv("../WeatherPy_Output/WeatherPyOutput.csv", index=False)

weather_df.head()
```

# Scatter Plots:

```python
#Build Scatter Plot
plt.scatter(weather_df['Latitude'], weather_df['Temperature (F)'],marker="o", facecolors="blue", edgecolors="black",
           label='City',alpha=0.75)
plt.title("Temperature (F) vs Latitude")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()

# Save the figure
plt.savefig("../Plot_Images/Latitude_Temperature.png")
```
